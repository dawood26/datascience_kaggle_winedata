```python
#Creating my perfect mask - I want Us,Israel and Germany but price < 20 dollars
mask_countries = (df.country == 'US') & (df.price < 20.0)| (df.country == 'Germany') & (df.price < 20.0) | (df.country == 'France') & (df.price < 20.0) | (df.country == 'Israel')& (df.price < 20.0) 
```

```python
wine_less_than_20=df.loc[mask_countries , : ]
```
